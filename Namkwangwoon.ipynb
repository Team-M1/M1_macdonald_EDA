```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./menu.csv')

# df = df.drop(['Category', 'Item', 'Serving Size'], axis=1)

df
```

```python
new_df = df.groupby('Category').mean()

new_df
```

```python
new_df.columns
```

```python
new_df = new_df[['Calories', 'Total Fat', 'Cholesterol', 'Sodium', 'Sugars', 'Protein']]

new_df
```

```python

plt.xticks(rotation = 75)
sns.barplot(
    data=new_df,
    x=new_df.index,
    y='Total Fat'
)
```

```python
fig, ax = plt.subplots(len(new_df.columns),1, figsize=(15, 25))
for i in range(len(ax)):
    sns.barplot(x=new_df.index, y=new_df.columns[i], data=new_df, ax=ax[i])
```
